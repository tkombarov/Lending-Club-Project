# Data Cleaning and Exploration

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "svg"

from pandas.tseries.offsets import *
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_excel('2007_2011.xlsx')
```

# Data Cleaning

As with any other dataset, first thing we want to do is to check for missing values. 
'missingo' is a great package that helps with visualizing missing data. 

```python
import missingno as msno
%matplotlib inline

columns_before_drop = len(df.columns.to_list())
```

```python
msno.matrix(df)
```

Above plot shows that majority of the columns from the original file are fully empty. 
Let's remove all columns where more than 50% of the values are missing.

```python
df = df.dropna(thresh=df.shape[0]*0.5,how='all',axis=1)
columns_after_drop = len(df.columns.to_list())
```

```python
round((columns_before_drop - columns_after_drop) / columns_before_drop, 2)
```

From the above, you can see that roughly 2/3 of the columns were missing more than 50% of the values. 

```python
msno.matrix(df)
```

missingo package helped us visualize obviuosly empty columns. What about the remaining columns ?                      
For each column, let's look into what % of rows has missing data and rank them in descending order.

This will help up to be more precise with further data cleaning. 

```python
percent_missing_by_column = round(df.isnull().sum() * 100 / len(df), 4)
percent_missing_by_column.sort_values(ascending=False)[:7]
```

Besides 'desc' and 'emp_title' columns, missing data is pretty under control. Let's refer to the data dictionary to undestand what this columns are about.

According to the data dictionary, 'desc' column is the loan proceeds description provided by the borrower. Almost 1/3 of this column is missing but we will leave it as is as we probably will not use it in the modeling (even though length of the description can potentially be a predictor).

Employment title column has the 2nd highest percentage of missing values but we won't be using this column either. 

Number of public record bankruptcies: 'pub_rec_bankruptcies' can already be an important variable. 

```python
del df['emp_title']
del df['desc']
```

Even though this is not the best practice, for simplicity let's remove all rows with any missing data.

```python
before_rows_drop = len(df.index)
```

```python
df = df.dropna(how='any',axis=0)
df.reset_index(inplace=True)
del df['index']
```

```python
after_rows_drop = len(df.index)
round((before_rows_drop - after_rows_drop) / before_rows_drop, 2)
```

Dataset shrunk by 6% which is significant but let's consider it acceptable. 

```python
df.head()
```

# Origination Trends

Below section breaksdown and looks on the trends in origination such as:
- Growth of the loan issuance over time
- Breakdown by loan term 
- Grade distribution and interest rates
- Proceeds purpose
- Geographical Distribution

# Origination by Year

```python
df['Issue_Year'] = pd.DatetimeIndex(df['issue_d']).year
```

```python
origination_by_year = round(df.groupby('Issue_Year')['loan_amnt'].sum(), 0)
fig = px.bar(origination_by_year, y=origination_by_year.values)
fig.update_layout(xaxis_title="Issue Year", yaxis_title="Dollars Originated")
fig.update_traces(marker_color='#EF4123')
fig.show()
```

From the above chart it is clear that vast majority of the origination has happend in 2011. In that year, Lending Club originated more than in all previous years.
- Something to keep in mind is that the economy started to take off again in 2011/2012 time period so valuable recession time origination data is very limited relative to the amount of issued loans. 

# Loan Term

```python
term_breakdown = round(df.groupby('term')['loan_amnt'].sum(), 0)

lc_colors = ['#EF4123', '#002A4E']

fig = px.pie(term_breakdown, values=term_breakdown.values, 
             names=term_breakdown.index, 
             title='Origination by Loan Term',
             color=term_breakdown.index,
             color_discrete_sequence=lc_colors)
fig.show()
```

# Loan Grades

```python
origination_by_grade = round(df.groupby('grade')['loan_amnt'].sum(), 0)
fig = px.pie(origination_by_grade, values=origination_by_grade.values, names=origination_by_grade.index, 
             title='Origination by Grade')
fig.show()
```

Almost half of the issued loans by dollar amount are graded B or above. At the same time there is a good proportion of C and D loans in the total amount issued between 2007 and 2011. This gives the spectrum of choice for different types of investors with the different risk appetite. 

Let's look into the interest rate charged for each grade level as well. 

```python
fig = px.box(df, x="grade", y="int_rate", category_orders={'grade':['A', 'B', 'C', 'D', 'E', 'F', 'G']})
fig.update_layout(xaxis_title="Grade", yaxis_title="Interest Rate")
fig.show()
```

As expected, higher the grade, higher the interest rate charged. Further analysis should include mean and not only median as well break down by the loan term. 

# Proceeds Purpose

```python
origination_by_purpose = round(df.groupby('purpose')['loan_amnt'].sum(), 0)
fig = px.pie(origination_by_purpose, values=origination_by_purpose.values, names=origination_by_purpose.index, 
             title='Origination by Purpose')
fig.show()
```

Roughly half of the loan proceeds are going towads debt consolidation which I would assume has to be bundled with the "credit card" category. Main source of revenue for Lending Club is refinancing the credit card debt at a less punitive interest rate. 

# Geographical Distribution

```python
origination_by_state = round(df.groupby('addr_state')['loan_amnt'].sum(), 0)

fig = go.Figure(data=go.Choropleth(
    locations=origination_by_state.index, 
    z = origination_by_state.values.astype(int), 
    locationmode = 'USA-states', 
    colorscale = 'YlGnBu',
    colorbar_title = "Millions USD",
))

fig.update_layout(
    title_text = '2007 - 2011 Lending Club Origination by State',
    geo_scope='usa', 
)

fig.show()
```

# Defaults 

One of the first things that a lot of investors would care about is the default rate. As mentioned above, dataset represents Lending Club loans originated between 2007 and 2011. Maturity of these loans is of 3 or 5 years so all the loans in the dataset have matured which is important as it doesn't create a censoring issue which is something that would be crucial in credit modeling part of the project. 

Let's slice and dice defaulted loans !

```python
df['loan_status'].value_counts()
```

Seems like the terminology issue or bad data, let's correct it. 

```python
df['loan_status'] = np.where(df['loan_status'] == 'Does not meet the credit policy. Status:Fully Paid', 
                             'Fully Paid', df['loan_status'])
```

```python
df['loan_status'] = np.where(df['loan_status'] == 'Does not meet the credit policy. Status:Charged Off', 
                             'Charged Off', df['loan_status'])
```

```python
loan_status = round(df.groupby('loan_status')['loan_amnt'].sum(), 0)

fig = px.pie(loan_status, values=loan_status.values, names=loan_status.index, title='Loan Status')
fig.show()
```

As we see from above, 15.5% of all loans have defaulted. It is hard to say whether it is a lot or not without a proper benchmark and without looking into the returns.

Something that is even more important is to analyze default rate by Grade assigned by Lending Club, potential investor should expect increase in default rate with the increase of the risk grade. If this assumption doesn't hold true then Lending Club's credit risk assessment would be questionable

```python
df['loan_status'] = np.where(df['loan_status'] == 'Charged Off', 1, 0)

defaults_by_grade = round(df.groupby('grade')['loan_status'].mean(), 4)
fig = px.bar(defaults_by_grade, y=defaults_by_grade.values)
fig.update_layout(xaxis_title="Grade", yaxis_title="Default Rate")
fig.show()
```

Default rate is increasing proportional to the risk grade, higher the risk, higher the default rate and vice versa. This shows that Lending Club actually prices loans according to their risk.

```python
defaults_by_year = round(df.groupby('Issue_Year')['loan_status'].mean(), 4)
fig = px.bar(defaults_by_year, y=defaults_by_year.values)
fig.update_layout(xaxis_title="Year", yaxis_title="Default Rate")
fig.show()
```

Unsurprisingly, 2007 and 2008 vintages have very high default rate as they got issued during or right before the recession. In addition, these are the first two vintages Lending Club ever issued so there was probably not enough data and processes to mitigate risk more properly.
